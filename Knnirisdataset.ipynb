```
I created this notebook as instructional material for the American High School AI club. 
from sklearn import neighbors, datasets
import pandas as pd

# load the data
# The iris data set labels 3 flower types by 4 different attributes of the flower (sepal length, sepal width, petal length, petal width).
data = datasets.load_iris()
```

This data we load is in the JSON format. We must next convert it into a 'dataframe' (a table) in order to work with it, using the pandas library

```
features=pd.DataFrame(data['data'],columns=data['feature_names'])
features.head() #see our dataframe (first 5 lines)
```

```
y = pd.Series(data['target'])#create a single column for our labels
y.head()
```

```
data['target_names'] ##above, 0 means a setosa, 1 means a versicolor, and 2 means a virginica
```

iris.jpeg

```
features.shape ##returns samples and columns (we have 150 samples and 4 columns)
```

To accurately test our model, we will divide our samples into training and testing data. We will keep 60% for training and 40% for testing.

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features,y,test_size=0.4,shuffle=True,stratify=y)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

Now, we have our features and target variables ready and we can train our model.

```
knn=neighbors.KNeighborsClassifier(n_neighbors=14) #create our model and set k=12
knn.fit(X_train, y_train)
knn.score(X_test,y_test) #tests our model and gives an accuracy score
```
